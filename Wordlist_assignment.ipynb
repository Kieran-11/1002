```python
wordlist = list(words) #casts words as a list "wordlist"

print("Question 1 solution: wordlist contains" , len(wordlist) , "strings") #prints length of wordlist
print("Question 2 solution:" , wordlist[0:9]) #prints first 10 of wordlist

alphalist = sorted(wordlist) #alphabatizes wordlist
print("Question 3 solution:",alphalist[0:9]) #prints first 10 of alphabatized list

a_words = 0
for word in wordlist: #iterates through wordlist checking the first letter of each word and incrementing a_words
  first_letter = word[0]
  if first_letter == "a":
    a_words += 1
print("Question 4 solution:" , a_words)

words_with_q = 0
for word in wordlist: #iterates through wordlist and checks each word for a "q" then increments words_with_q
  if "q" in word:
    words_with_q += 1
print("Question 5a solution:" , words_with_q)

words_without_o = 0
for word in wordlist: #uses opposite function of previous for loop to check for words wihtout o
  if "o" not in word:
    words_without_o +=1
print("Question 5b solution:" , words_without_o/len(wordlist))
```

```python
import numpy as np
import matplotlib.pyplot as plt
alphabet = [chr(i) for i in range(ord('a'),ord('z')+1)] #creates a list of the alphabet
first_letter_counter = np.zeros(26) #creates empty array size of alphabet that will fill with values of words that start with each letter

for word in wordlist: #increments the value of first counter array at the index of the first letter of each word
    first_letter = word[0]
    letter_index = ord(first_letter) - ord('a')
    if letter_index >= 0 and letter_index < 26:
        first_letter_counter[letter_index] += 1

print("Problem 7 solution:")
x1 = alphabet
y1 = first_letter_counter
plt.bar(x1, y1, color = 'g')
plt.plot
plt.xlabel("First Letter")
plt.ylabel("Word Count")
plt.title("Word Starting Letter Count")
```

```python
sorted_wordlist = sorted(wordlist, key=len, reverse=True) #function that sorts arrays by string length
longest_six = sorted_wordlist[:6] #finds first 6 strings in the array
print("Problem 8 solution:", longest_six)

longest_word_length = len(sorted_wordlist[0]) #finds longest word
word_length_index = np.arange(1, longest_word_length + 1) #creates an array from 1 to the longest word
word_length_counter = np.zeros(len(sorted_wordlist[0])) #creates array of zeros with size up to length og longest word

for word in wordlist: #goes through word list and increments the empty array at the index of the word length
  length = len(word)
  word_length_counter[length-1] += 1

print("Problem 9 solution:")
plt.bar(word_length_index, word_length_counter, color='b')
plt.xlabel('Word Length')
plt.ylabel('Number of Words')
plt.title('Histogram of Word Lengths')
plt.xticks(range(0, longest_word_length))
plt.show()
```
