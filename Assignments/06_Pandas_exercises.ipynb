10.  Load the `iris` dataset that we used in class.  Use `iris` as the name of your data frame.

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.iloc[50:100]
```

11.  What is the correlation between the variables `sepal_length` and `sepal_width` for all the irises?

```python
length = iris.sepal_length.values
width = iris.sepal_width.values
plt.scatter(length, width)
plt.xlabel('Length')
plt.ylabel('Width')
plt.show()
r = np.corrcoef(length, width)[0,1] #numpy function to find r value between two arrays
print('r value is', r)
```

12.  Which of the three species has the highest correlation between `sepal_length` and `sepal_width`?  Show any calculations you use to support your answer.

```python
#plots setosa sepal length vs sepal width and gives r value
setosa_sepal_length = iris['sepal_length'].head(50).values
setosa_sepal_width = iris['sepal_width'].head(50).values
r_setosa = np.corrcoef(setosa_sepal_length, setosa_sepal_width)[0,1]
print("setosa r value:", r_setosa)

#plots versicolor sepal length vs sepal width and gives r value
versicolor_sepal_length = iris['sepal_length'].iloc[50:100].values
versicolor_sepal_width = iris['sepal_width'].iloc[50:100].values
r_versicolor = np.corrcoef(versicolor_sepal_length, versicolor_sepal_width)[0,1]
print("versicolor r value:", r_versicolor)

#plots virginica sepal length vs sepal width and gives r value
virginica_sepal_length = iris['sepal_length'].tail(50).values
virginica_sepal_width = iris['sepal_width'].tail(50).values
r_virginica = np.corrcoef(virginica_sepal_length, virginica_sepal_width)[0,1]
print("virginica r value:", r_virginica)
print("")
lowest_r = min(r_setosa, r_versicolor, r_virginica) #gives smallest r value
print(lowest_r, "is the lowest r value")
```
