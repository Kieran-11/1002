```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import requests
from skimage import io

def red_filter(image_np):
  red_image = np.copy(image_np)
  red_image[:,:,1] = 0
  red_image[:,:,2] = 0
  return red_image

def blue_filter(image_np):
  blue_image = np.copy(image_np)
  blue_image[:,:,0] = 0
  blue_image[:,:,1] = 0
  return blue_image

def green_filter(image_np):
  green_image = np.copy(image_np)
  green_image[:,:,0] = 0
  green_image[:,:,2] = 0
  return green_image

#url of first image
#https://s3.amazonaws.com/eit-planttoolbox-prod/media/images/Bellis_Perennis--Morgaine--CC_BY_2.0.jpg
print("Please paste the URL of the jpeg you would like filtered")
url = input()
image = np.asarray(Image.open(requests.get(url, stream=True).raw))

#show original image
plt.imshow(image)
plt.axis('off')  # looked this up to get rid of axis around image
plt.show()

#application of filters to image
red_filter_image = red_filter(image)
green_filter_image = green_filter(image)
blue_filter_image = blue_filter(image)

#show red filtered image
plt.imshow(red_filter_image)
plt.axis('off')  # looked this up to get rid of axis around image
plt.show()

#show green filtered image
plt.imshow(green_filter_image)
plt.axis('off')  # turn off axis labels
plt.show()

#show blue filtered image
plt.imshow(blue_filter_image)
plt.axis('off')  # turn off axis labels
plt.show()

#mixed filter images
red_green_image = red_filter_image + green_filter_image
red_blue_image = red_filter_image + blue_filter_image
blue_green_image = blue_filter_image + green_filter_image

#show red green image
plt.imshow(red_green_image)
plt.axis('off')  # turn off axis labels
plt.show()

#show red blue image
plt.imshow(red_blue_image)
plt.axis('off')  # turn off axis labels
plt.show()

#show blue green image
plt.imshow(blue_green_image)
plt.axis('off')  # turn off axis labels
plt.show()

if np.array_equal(image, red_filter_image + green_filter_image + blue_filter_image): #numpy function to check if two arrays are equal
  print("They match!")
else:
  print("They do not match.")
```
